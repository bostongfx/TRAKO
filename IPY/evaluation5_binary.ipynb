```python
import warnings
warnings.filterwarnings("ignore") # hide vtk.util warning
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()

data = [os.path.getsize(v) for v in ['../DATA/example.vtp','../DATA/example.tko','../DATA/example.glb']]
names = ['Original', 'TKO', 'binary TKO']
    
fig.add_trace(go.Bar(y=data[0:2], x=names[0:2],
                marker_color = 'lightslategray',    orientation='v')
)


fig.add_trace(go.Bar(y=data[2:], x=names[2:],
                marker_color = 'darkred',    orientation='v')
)


fig.update_layout(yaxis_title="File Size")
# fig.update_layout(xaxis_title="Mean Relative Loss per Value")
    
# fig.update_layout(xaxis_type="log",xaxis=dict(tickformat=".5f"))
fig.show()
```
